```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def make_df(data):
    data_list = []
    for line in data.strip().split('\n'):
        values = line.split(',')
        data_dict = {}
        for value in values:
            key, val = value.strip().split(': ')
            data_dict[key] = float(val)
        data_list.append(data_dict)

    # Create a DataFrame
    return pd.DataFrame(data_list)
```

```python
plt.plot([ 1, 4, 7, 10, 13, 15], exp2['# of packet in system'].values, marker='o')
plt.grid()
plt.title('Experiement 2a: Population with varying packet arrival rate')
plt.xlabel('Packet Arrrival Rate')
plt.ylabel('Population')
plt.show()
```
