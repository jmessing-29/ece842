## Experiment 1

```python
#### Input Parameters

n = 2

# Arrival Rate: G = n*g, G= 0.5 1 1.5 2 2.5 3 3.5 4 4.5 5 5.5 6 = 2*g so g = [ varies ] / 2
g = np.arange(0.5, 6.5, 0.5)/2

rang = 10

time = 100000

file = 'exp1topo.txt'
```

```python
for rate in g:
    cmd = ('ALOHAsim.exe ' + str(n) + ' ' + str(rate) + ' ' + str(rang) + ' ' + str(time) + ' ' + file)
    ! $cmd
    ! echo. >> exp1res.txt
    ! echo. >> exp1res.txt
    ! echo. ############## experimemnt break: g = $rate ################# >> exp1res.txt
    ! type results.txt >> exp1res.txt
```
